```python
red['wine_type'] = 0
white['wine_type'] = 1

data = pd.concat([red,white], ignore_index= True)
data.head()
```

```python
X = data.drop(columns=['wine_type','quality'])
y = data['wine_type']
```

```python
from sklearn.model_selection import train_test_split 
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)
```

```python
knn = KNeighborsClassifier(n_neighbors=5)  # Using k=5 initially
knn.fit(X_train, y_train)
```

```python
y_pred = knn.predict(X_test)
print("Accuracy:", accuracy_score(y_test, y_pred))
print("\nClassification Report:\n", classification_report(y_test, y_pred))
print("\nConfusion Matrix:\n", confusion_matrix(y_test, y_pred))
```

```python
import seaborn as sns
from sklearn.decomposition import PCA
```

```python
# Boxplot of Key Features
plt.figure(figsize=(12, 8))
features_to_plot = ['alcohol', 'volatile acidity', 'citric acid', 'density']
for i, feature in enumerate(features_to_plot, 1):
    plt.subplot(2, 2, i)
    sns.boxplot(x='wine_type', y=feature, data=data)
    plt.title(f'{feature} by Wine Type')
plt.tight_layout()
plt.show()
```

```python
# Pair Plot to See Feature Relationships

sns.pairplot(data, hue='wine_type', vars=['alcohol', 'volatile acidity', 'citric acid', 'density'])
plt.show()
```

```python
# PCA Visualization
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)  # X_scaled from previous code
data_pca = pd.DataFrame(X_pca, columns=['PCA1', 'PCA2'])
data_pca['wine_type'] = y.replace({0: 'red', 1: 'white'})
```

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(x='PCA1', y='PCA2', hue='wine_type', data=data_pca, alpha=0.7)
plt.title('PCA: Red vs White Wine')
plt.show()
```
